```python
import numpy as np
from sklearn.cluster import KMeans

n = int(input())
k = int(input())
X = []
for _ in range(n):
    X.append(list(map(float, input().split())))

data = np.array(X)
model = KMeans(n_clusters=k, random_state=0, n_init=10)
model.fit(data)

# Remove round() — let %.2f handle the formatting directly
print("%.2f" % model.inertia_)
```
